Here are the instructions to setup multinode cluster with auto scaling using Databricks.
- If you are using Azure, make sure quota for desired worker and master type is raised to 60.
- Goto Compute and click on Create Cluster
- Choose Multi Node (which is default)
- Make sure to choose latest run time
- Choose Worker Type (Close to 16 GB RAM and 4 Cores)
- Make sure to choose 2 as Minimum and 8 as Maximum
- Choose Driver Type same as Worker Type 
- Configure Auto Termination Time to desired value (preferably 15 minutes)
- Wait until the cluster is created
