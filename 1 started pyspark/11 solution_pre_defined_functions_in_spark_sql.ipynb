# Answer 1

```
%sql

select (date_format(create_ts, 'yyyy')) as year
,count(1)
from users
group by 1
order by 1
```

# Answer 2

```
%sql

select user_id,
user_dob,
user_email_id,
date_format(user_dob, 'EEEE') as user_day_of_birth
from users
```

```
%sql

select user_id,
user_dob,
user_email_id,
date_format(user_dob, 'EEEE') as user_day_of_birth
from users
where month(user_dob) = 5
```

# Answer 3

```
%sql

select user_id,
upper(concat(user_first_name,' ',user_last_name )) as user_name,
user_email_id,
create_ts,
year(create_ts) as created_year
from users
where year(create_ts) = 2019
```

# Answer 4

```
%sql

select  case
            when user_gender = 'M' then 'Male'
            when user_gender = 'F' then 'Female'
            else 'Not Specified'
        end user_gender
        ,count(1) as user_count
from users
group by 1
order by 2 desc
```

# Answer 5

```
%sql

select user_id,
user_unique_id,
case
  when length(replace(user_unique_id, '-', '')) = 9 then substring(replace(user_unique_id, '-', ''), -4)
  when user_unique_id is null then 'Not Specified'
  else 'Invalid Unique Id'
end as user_unique_id_last4
from users
```

# Answer 8

```
%sql

select order_id,
order_date,
dayofweek(order_date),
date_format(order_date, 'EE')
from orders
limit 10
```

```
%sql

select
  case
    when date_format(order_date, 'EE') in ('Sun', 'Sat') then 'Weekend days'
    else 'Week days'
  end as day_type,
  count(*) as order_count
from orders
group by 1
order by 2
```

```
%sql

select
  case
    when dayofweek(order_date) in (1,7) then 'Weekend days'
    else 'Week days'
  end as day_type,
  count(*) as order_count
from orders
where date_format(order_date, 'yyyyMM') = 201401
group by 1
order by 2
```
