# Using CASE and WHEN in Spark SQL

```
%sql
use itversity_retail
```

```
%sql
describe orders
```

```
%sql
select distinct order_status
from orders
```

```
%sql

select  o.order_status,
        case
          when o.order_status in ('COMPLETE','CLOSED') then 'COMPLETE'
          when o.order_status in ('PENDING','PENDING_PAYMENT','PROCESSING','PAYMENT_REVIEW') then 'PENDING'
          else 'OTHER'
        end updated_order_status
from orders o
```

```
%sql

select  distinct o.order_status,
        case
          when o.order_status in ('COMPLETE','CLOSED') then 'COMPLETE'
          when o.order_status in ('PENDING','PENDING_PAYMENT','PROCESSING','PAYMENT_REVIEW') then 'PENDING'
          else 'OTHER'
        end updated_order_status
from orders o
```

```
%sql

select  case
          when o.order_status in ('COMPLETE','CLOSED') then 'COMPLETE'
          when o.order_status in ('PENDING','PENDING_PAYMENT','PROCESSING','PAYMENT_REVIEW') then 'PENDING'
          else 'OTHER'
        end updated_order_status
        ,count(*) as order_count
from orders o
group by 1
order by 1
```
